# QQ ploty

```python
import scipy.stats as stats

plt.figure(figsize=(8, 4))
stats.probplot(df[df['gender'] == 'female']['age'], dist="norm", plot=plt)
plt.title('QQ plot wieku - Kobiety')
plt.xlabel('Teoretyczne kwantyle')
plt.ylabel('Empiryczne kwantyle')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 4))
stats.probplot(df[df['gender'] == 'male']['age'], dist="norm", plot=plt)
plt.title('QQ plot wieku - Mężczyźni')
plt.xlabel('Teoretyczne kwantyle')
plt.ylabel('Empiryczne kwantyle')
plt.grid(True)
plt.show()
```
